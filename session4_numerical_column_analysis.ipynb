```python
pwage=visa_df['prevailing_wage']
Q1=np.percentile(pwage,25)
Q2=np.percentile(pwage,50)
Q3=np.percentile(pwage,75)
iqr=Q3-Q1
lb=Q1-1.5*iqr
ub=Q3+1.5*iqr
con1=pwage<lb
con2=pwage>ub
con=con1 | con2
outliers_df=visa_df[con]
outliers_df['prevailing_wage'].values
outliers_list=outliers_df['prevailing_wage'].values.tolist()
```

```python
Median=visa_df['prevailing_wage'].median()
new_wdata=[]
for i in pwage:
    if i in outliers_list:
        new_wdata.append(Median)
    else:
        new_wdata.append(i)
len(new_wdata)
        
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.hist(pwage)
plt.subplot(2,2,2)
plt.hist(new_wdata)
plt.subplot(2,2,3)
plt.boxplot(pwage,vert=False)
plt.subplot(2,2,4)
plt.boxplot(new_wdata,vert=False)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1).hist(pwage)
plt.subplot(2,2,2).hist(new_wdata)
plt.subplot(2,2,3).boxplot(pwage,vert=False)
plt.subplot(2,2,4).boxplot(new_wdata,vert=False)
plt.show()
```

**winsorization**

- whatever the outliers less than lb fill with lb

- whatever the outliers greater than ub fill with ub

- in numpy we have method **clip**

```python
pwage=visa_df['prevailing_wage']
Q1=np.percentile(pwage,25)
Q2=np.percentile(pwage,50)
Q3=np.percentile(pwage,75)
iqr=Q3-Q1
lb=Q1-1.5*iqr
ub=Q3+1.5*iqr
wage_dclip=np.clip(pwage,lb,ub)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(3,2,1).hist(pwage)
plt.subplot(3,2,2).hist(new_wdata)
plt.subplot(3,2,5).hist(wage_dclip)
plt.subplot(3,2,3).boxplot(pwage,vert=False)
plt.subplot(3,2,4).boxplot(new_wdata,vert=False)
plt.subplot(3,2,6).boxplot(wage_dclip,vert=False)
plt.show()
```
